```python
# Install data
install_data = {
    'Month': ['July', 'August', 'September'],
    'USA': ['10k', '8k', '6k'],
    'UK': ['7k', '5k', '3k'],
    'Australia': ['5k', '4k', '3k'],
    'India': ['9k', '7k', '5k'],
    'Brazil': ['6k', '5k', '4k']
}

# DAU data
dau_data = {
    'Month':["September", "September","September"],
    'Date': ['Sept 1', 'Sept 15', 'Sept 30'],
    'DAU': ['32k', '28k', '24k']
}

# MAU data
mau_data = {
    'Month': ['July', 'August', 'September'],
    'MAU': ['195k', '170k', '145k']
}

# Average Session Duration
avg_session_duration = {
    'Month': ['July', 'August', 'September'],
    'Avg_Session_Duration': ['7 minutes 30 seconds', '5 minutes 45 seconds', '4 minutes 10 seconds']
}

# Create dataframes
install_df = pd.DataFrame(install_data)
dau_df = pd.DataFrame(dau_data)
mau_df = pd.DataFrame(mau_data)
session_duration_df = pd.DataFrame(avg_session_duration)
```

```python
screen_flow_data = {
    'Screen Name': ['Home', 'Search', 'Booking', 'Payment', 'Confirmation'],
    'Visits': ['70%', '50%', '30%', '10%', '2%'],
    'Drop-Off Rate': ['30%', '25%', '50%', '80%', '0%']
}

# Retention Data
retention_data = {
    'Month': ['July', 'August', 'September'],
    'Weekly_retention': ['74%', '65%', '55%'],
    'Monthly_retention': ['61%', '50%', '40%']
}

# Churn Data
churn_data = {
    'Month': ['July', 'August', 'September'],
    'Weekly_churn': ['26%', '35%', '45%'],
    'Monthly_churn': ['39%', '50%', '60%']
}

# In-App Purchases Data
in_app_purchases_data = {
    'Month': ['July', 'August', 'September'],
    'Premium Features': ['$5.2k (2%)', '$4k (1.5%)', '$2.5k (1%)'],
    'Travel Guides': ['$3.1k (1.5%)', '$2.5k (1.2%)', '$1.5k (1%)']
}

screen_flow_df = pd.DataFrame(screen_flow_data)
retention_df = pd.DataFrame(retention_data)
churn_df = pd.DataFrame(churn_data)
```

```python
in_app_purchases_df=pd.DataFrame(in_app_purchases_data)
```

```python
ad_revenue_data = {
    'Month': ['July', 'August', 'September'],
    'Impressions': ['980k', '850k', '750k'],
    'Click-Through': ['4.8%', '4.4%', '4%'],
    'Revenue': ['$9.8k', '$7.5k', '$5k']
}

# User Ratings Data
user_ratings_data = {
    'Month': ['July', 'August', 'September'],
    '5-Star': ['2.5k', '1.8k', '1k'],
    '4-Star': ['1.8k', '1.2k', '0.8k'],
    '3-Star': ['1.5k', '1k', '0.7k'],
    '2-Star': ['1k', '1.2k', '1.5k'],
    '1-Star': ['3.2k', '3.8k', '4.5k']
}

# App Crash Reports Data
app_crash_reports_data = {
    'Month': ['July', 'August', 'September'],
    'App Crash Reports': [140, 500, 400]
}

# Customer Support Tickets Data
support_tickets_data = {
    'Month': ['July', 'August', 'September'],
    'Customer Support Tickets': [410, 700, 300]
}

# Create DataFrames for the new data
ad_revenue_df = pd.DataFrame(ad_revenue_data)
user_ratings_df = pd.DataFrame(user_ratings_data)
app_crash_reports_df = pd.DataFrame(app_crash_reports_data)
support_tickets_df = pd.DataFrame(support_tickets_data)
```

```python
merged_df = pd.merge(install_df, mau_df, on='Month', how='outer')

# Merge with session_duration
merged_df = pd.merge(merged_df, session_duration_df, on='Month', how='outer')

# Merge Retention data
merged_df = pd.merge(merged_df, retention_df, on='Month', how='outer')

# Merge Churn data
merged_df = pd.merge(merged_df, churn_df, on='Month', how='outer')

# Merge In-App Purchases data
merged_df = pd.merge(merged_df, in_app_purchases_df, on='Month', how='outer')

# Merge Ad Revenue data
merged_df = pd.merge(merged_df, ad_revenue_df, on='Month', how='outer')

# Merge User Ratings data
merged_df = pd.merge(merged_df, user_ratings_df, on='Month', how='outer')

# Merge App Crash Reports data
merged_df = pd.merge(merged_df, app_crash_reports_df, on='Month', how='outer')

# Merge Customer Support Tickets data
merged_df = pd.merge(merged_df, support_tickets_df, on='Month', how='outer')
```

```python
def convert_to_int(value):
    if isinstance(value, str) and 'k' in value:
        return int(float(value.replace('k', '')) * 1000)
    elif isinstance(value, str) and '%' in value:
        return float(value.replace('%', '')) / 100
    elif isinstance(value, str):
        value = value.replace('$', '').replace('k', '')
        return float(value) * 1000
    return value
```

```python
merged_df=merged_df.apply(convert_to_int)
```

```python
def convertion(time_str):
    try:
        parts = time_str.split()
        minutes = int(parts[0])
        seconds = int(parts[2])
        return minutes + seconds / 60
    except Exception as e:
        print(f"Error: {e} with input {time_str}")
        return None

merged_df['Avg_Session_Duration']=merged_df['Avg_Session_Duration'].apply(convertion)
```
